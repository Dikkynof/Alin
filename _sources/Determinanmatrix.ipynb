```python
# importing Numpy package
import numpy as np

# creating a 3X3 Numpy matrix
n_array = np.array([[1, 0, 0],
					[0, 1, 0],
					[0, 0, 1]])

# Displaying the Matrix
print("Numpy Matrix is:")
print(n_array)

# calculating the determinant of matrix
det = np.linalg.det(n_array)

print("\nDeterminant of given 3X3 square matrix:")
print(int(det))
```

```python
# importing Numpy package
import numpy as np

# creating a 2X2 Numpy matrix
n_array = np.array([[1, 0], [0, 1]])

# Displaying the Matrix
print("Numpy Matrix is:")
print(n_array)

# calculating the determinant of matrix
det = np.linalg.det(n_array)

print("\nDeterminant of given 2X2 matrix:")
print(int(det))
```

### sifat 4 : Jika matrix tersebut adalah kelipatan sebanyak n * matrix maka untuk mencari determinan adalah (n^(sebanyak ordo)determinan matrix)contoh matrix g adalah kelipatan 2 kali dari matrix d : (2^3(det d))

```python
g = np.array([[6, 2, 4],
					[2, 4, 2],
					[4, 2, 4]])
print(g)
print("sifat 4 : Jika Matrix tersebut adalah kelipatan 2 x matrix d maka dapat disimpulkan (2^3(det d))")
det = np.linalg.det(g)
print(int(det))
```

#### Contoh dengan ordo 2x2

### Sifat 5 : Jika terdapat matrix baru dengan nilai salah satu baris atau kolom adalah kelipatan 2 dari matrix sebelumnya maka dapat disimpulkan dengan 2*(determinan matrix)
contoh matrix i baris pertama adalah kelipatan 2 dari matrix a

```python
i = np.array([[6, 8],
					[2, 1]])
print(i)
print("sifat 5 : jika terdapat matrix baru dengan nilai salah satu baris adalah kelipatan 2 dari matrix sebelumnya maka dapat disimpulkan dengan 2x(det a)")
det = np.linalg.det(i)
print(det.round())
```

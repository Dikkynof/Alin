```python
## Aljabar Matrix

- Berikan penjelasan tentang apa itu Matrix

- Operasi Matrix

- Penjumlahan perkalian, pengurangan, transpose matrix

- Buat program untuk mengalikan matrix A dan B
```

```python
import numpy as np
```

```python
#Contoh Matrix
A = np.array([[8,9,3],[1,2,5],[2,1,1]])
B = np.array([[2,6,4],[1,2,3],[2,5,5]])
C = np.array([[9,2],[2,5],[4,0]])
D = np.array([[1,8,6],[0,5,3],[1,9,2]])

print("A= \n", A)
print("B= \n", B)
print("C= \n", C)
print("D= \n", D)
```

```python
A = np.array([[2,3,4],[5,6,7],[8,9,1]])
print("A = \n", A)
print()

B = np.array([[5,2,4],[3,7,9],[2,2,4]])
print("B = \n", B)
print()

#operasi matrix
print("E = \n", A+B)
print("F = \n", 2*A)
print("G = \n", A@C)
```

```python
print(A@(B@C))
print((A@B)@C)

print("-"*40)

print(7*(A@B))
print((7*A)@B)
print(A@(7*B))
```

```python
C = np.array([[9,2],[2,5],[4,0]])
print("C = \n", C)
C_T = C.transpose()
print(C_T)
print("-"*30)
print(C_T@C)
print("-"*30)
print(C@C_T)
```

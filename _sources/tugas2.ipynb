## Pengertian Matrix
Matrix adalah susunan bilangan real atau bilangan kompleks (atau elemen-elemen) yang disusun dalam baris dan kolom sehingga membentuk jajaran persegi panjang.

Matrix terdiri dari beberapa jenis sebagai berikut :

*  Matrix persegi
*  Matrix kolom
*  Matrix baris
*  Matrix Transpose

Operasi Matrix

Operasi Matrix adalah operasi yang dilakukan terhadap sebuah satu atau lebih matrix .Operasi matrix pada dasarnya meliputi : penjumlahan, pengurangan, perkalian.

1.  Penjumlahan matrix <br>
penjumlahan matrix adalah penjumlahan yang dilakukan terhadap dua buah matrix dengan komponen=komponennya yang seletak. Matrix ini dapat dijumlahkan jika jumlah baris dan kolomnya sama dan juga hasil dari penjumlahan ini juga akan memiliki ordo yang sama.

```python
import numpy as np
```

```python
# Contoh Matrix

A = np.array([[1,4,8],[2,-2,5],[7,3,1]]) # matrix 3x3
B = np.array([[2,6,3],[3,4,4],[2,-1,6]]) # matrix 3x3
C = np.array([[3,5],[-2,3],[4,3]]) # matrix 3x2
D = np.array([[2,5,8],[2,-4,5],[-3,5,2]]) # matrix 3x3
E = (A+B)
F = 2*(A)
G = A @ C
```

### Operasi Perkalian Matrix

```python
# Matrix A x C
H = A @ C
print(f"Matrix H =\n{H}")
print(f"Ukuran Matrix = {H.shape}")
```

### Tranpose Matrix

Anggap Baris = m, Kolom = n

C = A_trans @ A

(n x m) @ (m x n) = n * n

contoh: [2 x 3] @ [3 x 2] = [2 x 2]

C = A @ A_trans

(m x n) @ (n x m) = m x m

contoh: [3 x 2] @ [2 x 3] = [3 x 3]

```python
# Contoh dengan Martrix 3x3

A_tran = A.transpose()
print('Matrik A sebelum di transpose=\n', A)
print()
print('Matrik A setelah di transpose=\n',A_tran)
print('Ordo A_transpose adalah',A_tran.shape)
Z = A_tran @ A
print()
print('Z = A_transpose x A =\n',Z)
print('Ordo Z adalah',Z.shape)
print()
P=A @ A_tran
print('P = A_transpose x A =\n',P)
print('Ordo P adalah',P.shape)
```
